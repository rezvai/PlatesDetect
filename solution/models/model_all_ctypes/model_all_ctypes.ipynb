## 1. Обучение модели

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov10n.pt')

model.train(data='data.yaml', epochs=50, optimizer='AdamW', plots=True, save=True, val=True, name='all_ctypes_model_v10n')
```

_____________________

## 2. Работа с полученной моделью и разметка

```python
import json 
import os

model_best_pt = YOLO(r'runs\detect\all_ctypes_model_v10n\weights\best.pt')

data_links = [r'..\..\al-ctype-1_meta.json',
        r'..\..\al-ctype-2_meta.json',
        r'..\..\al-ctype-3_meta.json',
        r'..\..\al-ctype-4_meta.json']

ctype_names = ['al-ctype-1_meta.json', 'al-ctype-2_meta.json', 'al-ctype-3_meta.json', 'al-ctype-4_meta.json']

ctypes = ['al-ctype-1', 'al-ctype-2', 'al-ctype-3', 'al-ctype-4']
count = 0

for link in data_links:
    with open(link, 'r') as f:
        data = json.load(f)
        for obj in data:
            for file in os.listdir(os.path.join('..\\..\\', ctypes[count], data[obj]['ads_link'].replace('/', '_'))):
                if file.startswith("real_") and file.endswith('.jpg'):
                    predict = model_best_pt.predict(os.path.join('..\\..\\', ctypes[count], data[obj]['ads_link'].replace('/', '_'), file))
                    if len(predict[0].boxes.xywh) > 0:
                        data[obj]['bbox'] = " ".join(list(map(str, list(predict[0].boxes.xywh[0].numpy()))))
                    else:
                        data[obj]['bbox'] = 'No detections'
        with open(f'..\\..\\{ctype_names[count]}', 'w') as fp:
            json.dump(data, fp)
    count += 1
```

Данные успешно размечены, вся разметка помещененна в поле "bbox" для каждого объекта.
